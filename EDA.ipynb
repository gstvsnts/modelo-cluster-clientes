Aqui é possível perceber que as colunas *Z_CostContact* e *Z_Revenue* possuem valor constante, o que não é útil para análise.

Prossigo com a remoção das colunas citadas:

Os próximos passos no tratamento dos dados serão:

- Unir as colunas *Kidhome* e *Teenhome* em uma só

- Tratamento da variável *DT_Customer* que é um dado de data em formato object

```python
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'], dayfirst = True)
```

```python
df['Kidhome'] = df['Kidhome'] + df['Teenhome']
```

```python
df.drop(columns = ['Teenhome'], inplace = True)
```

```python
df.rename(columns = {'Kidhome':'Children'}, inplace = True)
```

```python
df.info()
```

```python
catcols = df.select_dtypes(include=['object']).columns.tolist()
numcols = df.drop(columns = 'ID').select_dtypes(include=['int64', 'float64']).columns.tolist()
```

```python
numcols
```

```python
fig, ax = plt.subplots(nrows = 1, ncols = 2, figsize = (15,6))

for idx,column in enumerate(catcols):
    plt.subplot(1,2,idx+1)
    sns.histplot(x = column, data = df)
    plt.title("{} Distribution" .format(column))
    plt.xlabel('')
```

```python
sns.boxplot(data = df, y = 'Income')
```

```python
sns.boxplot(data = df, y = 'Year_Birth')
```

```python
df = df[df['Income'] < 150000]
```

```python
fig, ax = plt.subplots(nrows = 6, ncols = 4, figsize = (25,30))

for idx,column in enumerate(numcols):
    plt.subplot(6,4,idx+1)
    sns.histplot(x = column, data = df, kde = True)
    plt.title("{} Distribution" .format(column))
    plt.xlabel('')
```
